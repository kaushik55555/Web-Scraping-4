```python
Video
```

```python
import re
```

```python
def name(a):
    if re.search('\[.*', a):
        pos = re.search('\[.*', a).start()
  
        return a[1:pos-1]
  
    else:
        return a
```

```python
Video["Video Name"]=Video["Video Name"].apply(name)
```
